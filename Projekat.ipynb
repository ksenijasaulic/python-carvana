```python
def calculate_scores(outcome, predictions,  message, train=False,):
    '''  Util function to calculate all scores and '''
    print(message)
    print("Scores measured on test data")
    print(f"Accuracy score: {accuracy_score(outcome,predictions)}")
    print(f"Precision score: {precision_score(outcome,predictions)}")
    print(f"Roc auc score: {roc_auc_score(outcome,predictions)}")
    print(f"Recall score: {recall_score(outcome, predictions)}")
    print()
```

### Errors Cost Matrix

```python
cost_matrix = np.matrix('0 10 ; 1 0')  

print('Matrica cena gresaka: ')
print(cost_matrix)

print('Napravljene greske: ')
print(confusion_matrix(model.predict(X),y))

costs = np.multiply(confusion_matrix(model.predict(X),y) , cost_matrix)
print('Cena napravljenih gresaka: ')
print(costs)

print('Ukupna cena napravljenih gresaka: ')
total_cost = (confusion_matrix(model.predict(X),y) * cost_matrix).sum().sum()
print(total_cost)

#Pravimo gresku tako sto predvidjamo dobra auta za ona koja su zapravo losa..
```

```python
threshold = 0.15
predictions = (model.predict_proba(X)[:,1] > threshold)     # poredjenje verovatnoca klase (True), sa definisanom granicom
predictions

print(accuracy_score(predictions,y))

print(f"Accuracy score: {accuracy_score(y, predictions)}")
print(f"Precision score: {precision_score(y, predictions)}")
print(f"Roc auc score: {roc_auc_score(y, predictions)}")
print(f"Recall score: {recall_score(y, predictions)}")


print(confusion_matrix(predictions,y))
print(classification_report(predictions,y))
                            
# povecavamo FN kako smanjujemo FP    
#Treba napomenuti da ova tehnika ne menja način rada algoritma, već samo utiče na to kako izračunate verovatnoće će biti pretvorene u klase koje model predviđa.                            
```
